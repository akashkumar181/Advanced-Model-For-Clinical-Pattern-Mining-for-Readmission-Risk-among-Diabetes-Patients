Drug status values (â€˜Noâ€™, â€˜Steadyâ€™, â€˜Upâ€™, â€˜Downâ€™) were converted into binary indicator variables per drug using one-hot encoding, allowing the ML model to interpret medication type and dose changes as separate feature

```python
drug_columns = [
    'metformin','insulin', 'glipizide', 'glyburide','pioglitazone',
    'rosiglitazone','acarbose','miglitol','troglitazone','tolbutamide',
    'chlorpropamide','glimepiride','glyburide-metformin',
    'metformin-rosiglitazone','metformin-pioglitazone'
]
```

```python
df = pd.get_dummies(df, columns=drug_columns, prefix=drug_columns, drop_first=True)
```

```python
df
```

```python
df['Drug_Count'] = df.filter(like='Up').sum(axis=1) + df.filter(like='Down').sum(axis=1)
```
