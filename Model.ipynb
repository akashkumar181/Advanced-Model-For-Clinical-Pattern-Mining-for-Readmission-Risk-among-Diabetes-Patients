#Diagnosis Report
1ï¸âƒ£ Diabetes and Heart Problems = Highest Re-Admission

People whose primary disease is diabetes (endocrine) or heart-related problems (circulatory) are most likely to come back to hospital

Because these diseases are long-term and need regular control

âœ” 2ï¸âƒ£ Early readmission (<30 days) is mainly because of health instability

Those who returned within 30 days, mostly had diabetes or heart issues

Meaning their condition was not stable even after discharge

Example: uncontrolled sugar, heart failure, BP fluctuating etc.

âœ” 3ï¸âƒ£ These two diseases are the biggest clinical burden

Both diabetes and cardiovascular diseases require continuous monitoring

They cause complications if not controlled

Hence more repeated admissions

âœ” 4ï¸âƒ£ Musculoskeletal (bones, joints) and Skin diseases = lowest readmission

These are less serious or chronic conditions

So people usually donâ€™t get admitted again very soon
5ï¸âƒ£ The pattern continues across Primary, Secondary AND Tertiary diagnosis

Meaning:

First diagnosis â†’ high readmission

Second diagnosis â†’ still high

Third diagnosis â†’ still high

So diabetes & heart conditions remain dominant across multiple diagnosis levels

âœ” 6ï¸âƒ£ Indicates multi-disease patients are more vulnerable

Many patients with diabetes also have:

hypertension

heart disease

kidney issues

->*Diabetes and heart diseases are the major medical reasons behind repeated hospital visits, especially within 30 days of discharge.*

```python
df['Diagnosis1'] = df['Diagnosis1'].astype(str)
df['Diagnosis2'] = df['Diagnosis2'].astype(str)
df['Diagnosis3'] = df['Diagnosis3'].astype(str)
```

```python
def map_icd(code):
    try:
        code = float(code)
    except:
        return 'Unknown'

    if 390 <= code <= 459:
        return 'Circulatory'
    elif 460 <= code <= 519:
        return 'Respiratory'
    elif 520 <= code <= 579:
        return 'Digestive'
    elif 580 <= code <= 629:
        return 'Genitourinary'
    elif 680 <= code <= 709:
        return 'Skin'
    elif 710 <= code <= 739:
        return 'Musculoskeletal'
    elif 780 <= code <= 799:
        return 'Symptoms'
    elif 240 <= code <= 279:
        return 'Endocrine'
    elif 140 <= code <= 239:
        return 'Cancer'
    elif 1 <= code <= 139:
        return 'Infection'
    elif 800 <= code <= 999:
        return 'Injury'
    else:
        return 'Other'
```

```python
df['PrimaryDisease'] = df['Diagnosis1'].apply(map_icd)
df['SecondaryDisease'] = df['Diagnosis2'].apply(map_icd)
df['TertiaryDisease'] = df['Diagnosis3'].apply(map_icd)
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = [
 'DaysInHospital','LabTests','Procedures','Medications',
 'PrevOutpatientVisits','PrevEmergencyVisits',
 'PrevInpatientVisits','DiagnosisCount'
]

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
X = df.drop('Readmission', axis=1)
y = df['Readmission']

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y,
                test_size=0.25, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

preds = model.predict(X_test)
print(accuracy_score(y_test, preds))
```
